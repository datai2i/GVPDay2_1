# This code covers various concepts such as creating line plots, scatter plots, bar plots, histograms, box plots, and pie charts using the Matplotlib library. Each section is accompanied by comments explaining the purpose and functionality of the code.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot

```python
# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Create a line plot
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Scatter Plot

```python
# Create data
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)
```

```python
# Create a scatter plot
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Bar Plot

```python
# Create data
x = ['A', 'B', 'C', 'D']
y = [10, 7, 5, 8]
```

```python
# Create a bar plot
plt.bar(x, y)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Histogram

```python
# Create data
data = np.random.randn(1000)
```

```python
# Create a histogram
plt.hist(data, bins=30)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

# Box Plot

```python
# Create data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

```python
# Create a box plot
plt.boxplot(data, labels=['A', 'B', 'C'])
plt.title("Box Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Pie Chart

```python
# Create data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
```

```python
# Create a pie chart
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title("Pie Chart")
plt.show()
```
